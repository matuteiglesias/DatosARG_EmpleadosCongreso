```python
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline

```

```python
x = pd.read_json('./datastore_search.json')
```

```python
# x.loc['fields']['result']
```

```python
x.index
```

```python
personal = pd.DataFrame(x.loc['records']['result'])

personal.nunique()
```

```python
personal.groupby('convenio').count()['_id']
```

```python
escalafon_ref = personal[['escalafon']].drop_duplicates()
escalafon_ref['A'] = escalafon_ref['categoria'] = escalafon_ref['convenio_id'] = ''

for i, row in escalafon_ref.iterrows():
    l = escalafon_ref['escalafon'].str.split('-')[i]
    escalafon_ref.loc[i,'A'] = l[0]; 
    escalafon_ref.loc[i,'categoria'] = l[1].zfill(2); 
    escalafon_ref.loc[i,'convenio_id'] = l[2]
    
    
personal = personal.merge(escalafon_ref)#
```

```python
cat_count = pd.DataFrame(personal.groupby('categoria').count()['_id']).reset_index()

df = pd.read_csv('./escala-salarial-sueldo-total.csv')
df['categoria'] = df['categoria'].astype(str).str.zfill(2)


# df.append({'categoria': 0, 'pesos': 0 }, ignore_index = True).style.bar(subset=['pesos'],color=['#5fba7d'])
df.merge(cat_count).style.bar(subset=['pesos', '_id'], color=['#d65f5f', '#5fba7d'])
```
